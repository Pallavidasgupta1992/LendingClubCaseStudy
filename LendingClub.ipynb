```python
#FUNCTION for diagonising blanks in the dataset

def missing_diag(dataframe, axis=1):

    if axis == 1:
    
        col = data.isnull().sum()
        col_percent = (col/len(data))*100
        df = pd.concat([col, col_percent], axis=1)
        df = df.rename(columns = {0 : 'Missing Values', 1 : 'Percent Missing Values'})
        df = df[df.iloc[:,1] != 0].sort_values('Percent Missing Values', ascending=False).round(1)
        print ("Dataset contains " + str(data.shape[1]) + " columns.\n"
               "There are " + str(df.shape[0]) + " columns that have missing values.\n")

        with pd.option_context('display.max_rows', None, 'display.max_columns', None):
            print(df)
        
        #Visualization of columns with missing values more than 30%
        col_percent_gt30 = col_percent[col_percent.values >(30)]
        plt.figure(figsize=(20,4))
        col_percent_gt30.plot(kind='bar')
        plt.title(str(len(col_percent_gt30))+ " columns with missing values more than 30%")
        plt.show()
            
    else :
        
        row = data.isnull().sum(axis = 1)
        row_percent = (row/data.shape[1])*100
        df = pd.concat([row, row_percent], axis=1)
        df = df.rename(columns = {0 : 'Missing Values', 1 : 'Percent Missing Values'})
        df = df[df.iloc[:,1] != 0].sort_values('Percent Missing Values', ascending=False).round(1)
        print ("Dataset contains " + str(data.shape[0]) + " rows.\n"
               "There are " + str(df.shape[0]) + " rows that have missing values.\n")
        print(df.head(10))

    return df
```

```python
#defining FUNCTION for dropping the columns/rows based on threshold values.
#remove_blanks(dataframe, axis, thresh) 

def remove_blanks(dataframe, axis=1, thresh=0.3):

    df = dataframe.copy()
    oldshape = df.shape
    if axis == 0:
        rows = df.transpose().isnull().sum()
        rows = list(rows[rows.values > thresh*len(df)].index)
        df.drop(df.index[rows],inplace=True) 
        print("\nNumber of rows dropped: ",len(rows))
    else:
        cols = (df.isnull().sum()/len(df))
        cols = list(cols[cols.values>=thresh].index)
        df.drop(labels = cols,axis =1,inplace=True)        
        print("Number of columns dropped: ",len(cols))
        
    print("\nOld Shape:",oldshape,"\nNew Shape:",df.shape)

    return df
```

```python
#Univariate function will plot the graphs based on ---

#df(dataframe), col(column), type_d(continuos(0) or categorical(1)), hue(only for categorical charts)
    

def univariate(df,col,type_d,hue = None):
    

    sns.set(style="darkgrid")
    
    if type_d == 0:
        fig, ax=plt.subplots(nrows =1,ncols=2,figsize=(20,8))
        ax[0].set_title("Distribution Plot")
        sns.distplot(df[col],ax=ax[0])
        ax[1].set_title("Box Plot")
        sns.boxplot(data =df, x=col,ax=ax[1],orient='v')
    
    if type_d == 1:
        temp = pd.Series(data = hue)
        fig, ax = plt.subplots()
        width = len(df[col].unique()) + 6 + 4*len(temp.unique())
        fig.set_size_inches(width , 7)
        ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue) 
        if len(temp.unique()) > 0:
            for p in ax.patches:
                ax.annotate('{:1.1f}%'.format((p.get_height()*100)/float(len(df_defaulters))), 
                            (p.get_x()+0.05, p.get_height()+20))  
        else:
            for p in ax.patches:
                ax.annotate(p.get_height(), (p.get_x()+0.32, p.get_height()+20)) 
        del temp
    else:
        exit
        
    plt.show()
```

```python
def crosstab(df,col):

    crosstab = pd.crosstab(df[col], df['loan_status'],margins=True)
    crosstab['Probability_Charged Off'] = round((crosstab['Charged Off']/crosstab['All']),3)
    crosstab = crosstab[0:-1]
    return crosstab
```

```python
def bivariate_prob(df,col,stacked= True):
    
    # get dataframe from crosstab function
    plotCrosstab = crosstab(df,col)
    
    linePlot = plotCrosstab[['Probability_Charged Off']]      
    barPlot =  plotCrosstab.iloc[:,0:2]
    ax = linePlot.plot(figsize=(20,8), marker='o',color = 'b')
    ax2 = barPlot.plot(kind='bar',ax = ax,rot=1,secondary_y=True,stacked=stacked)
    ax.set_title(df[col].name.title()+' vs Probability Charge Off',fontsize=20,weight="bold")
    ax.set_xlabel(df[col].name.title(),fontsize=14)
    ax.set_ylabel('Probability of Charged off',color = 'b',fontsize=14)
    ax2.set_ylabel('Number of Applicants',color = 'g',fontsize=14)
    plt.show()
```
